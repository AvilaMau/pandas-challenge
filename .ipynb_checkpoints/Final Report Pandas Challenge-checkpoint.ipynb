```python
# Dependencies and setup
import os
import csv
import pandas as pd
```

```python
# File to load
schoolpath = "../pandas-challenge/schools_complete.csv"
studentpath = "../pandas-challenge/students_complete.csv"
MIN_PASSING_SCORE = 70
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(schoolpath)
student_data = pd.read_csv(studentpath)
```

```python
# Combine the data into a single dataset.
school_data_total = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_total
```

# District Summary

```python
# Calculate the total number of unique schools
total_schools = school_data_total["school_name"].nunique()
school_data_total["school_name"].nunique()
```

```python
# Calculate the total number of students
student_count = school_data_total["student_name"].count()
school_data_total["student_name"].count()
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
school_data["budget"].sum()
```

```python
# Calculate the average (mean) math score
ave_math = student_data["math_score"].mean()
student_data["math_score"].mean()
```

```python
# Calculate the average (mean) reading score
ave_reading = student_data["reading_score"].mean()
student_data["reading_score"].mean()
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_total[(school_data_total["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated) 
passing_reading_count = school_data_total[(school_data_total["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_total[
    (school_data_total["math_score"] >= 70) & (school_data_total["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
                    "Total Schools":[total_schools], 
                    "Total Students":[student_count],
                    "Total Budget":[total_budget],
                    "Average Math Score":[ave_math],
                    "Average Reading Score":[ave_reading],
                    "% Passing Math":[passing_math_percentage],
                    "% Passing Reading":[passing_reading_percentage], 
                    "% Overall Passing":[overall_passing_rate]})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

# School Summary

```python
# Use the code provided to select the school type
school_types = school_data_total.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
per_school_counts =  school_data_total.set_index(["student_name"])["school_name"].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_total.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
per_school_capita, per_school_budget
```

```python
# Calculate the average test scores
per_school_math = school_data_total.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_total.groupby(["school_name"]).mean()["reading_score"]
per_school_reading, per_school_math
```

```python
# Calculate the number of schools with math scores of 70 or higher
#school_data_total.groupby(["school_name"])
school_passing_math = per_school_math[per_school_math >70]
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = per_school_reading[per_school_reading >70]
school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_total[
    (school_data_total["reading_score"] >= 70) & (school_data_total["math_score"] >= 70)
]
passing_math_and_reading
```

```python
passing_math_s = school_data_total["math_score"] >= MIN_PASSING_SCORE
passing_math_df = school_data_total[passing_math_s]
passing_reading_s = school_data_total["reading_score"] >= MIN_PASSING_SCORE
passing_reading_df = school_data_total[passing_reading_s]
passing_reading_df
```

```python
passing_math_reading_df = school_data_total[passing_math_s & passing_reading_s]
passing_math_reading_df
```

```python
# Use the provided code to calculate the passing rates
passing_math_rate_s = 100*(passing_math_df.groupby(["school_name"])["student_name"].count() / per_school_counts)
passing_reading_rate_s = 100*(passing_reading_df.groupby(["school_name"])["student_name"].count() / per_school_counts)
passing_math_reading_s = 100*(passing_math_reading_df.groupby(["school_name"])["student_name"].count() / per_school_counts)

passing_math_rate_s,passing_reading_rate_s,passing_math_reading_s
```
